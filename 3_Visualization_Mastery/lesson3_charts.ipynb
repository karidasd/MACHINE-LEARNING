# 🎨 Μάθημα 3: Οπτικοποίηση Δεδομένων (Data Visualization)
Σε αυτό το μάθημα θα μάθουμε να λέμε ιστορίες με τα δεδομένα μας χρησιμοποιώντας τις βιβλιοθήκες **Matplotlib** και **Seaborn**.

### Τι θα μάθουμε:
1. **Bar Charts**: Σύγκριση κατηγοριών.
2. **Line Charts**: Τάσεις στο χρόνο.
3. **Scatter Plots**: Σχέσεις μεταξύ μεταβλητών.
4. **Box Plots**: Εντοπισμός ακραίων τιμών (Outliers).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Φόρτωση των δεδομένων
df = pd.read_csv('../02_Datasets/sales_data.csv')
df['Συνολικό_Έσοδο'] = df['Τιμή'] * df['Ποσότητα']
print("✅ Δεδομένα έτοιμα για οπτικοποίηση!")
```

### 1️⃣ Σύγκριση Πωλήσεων (Bar Chart)
Ποιο προϊόν μας έφερε τα περισσότερα χρήματα;

```python
plt.figure(figsize=(10,5))
sns.set_style("whitegrid") # Προσθήκη πλέγματος στο background

chart = sns.barplot(x='Προϊόν', y='Συνολικό_Έσοδο', data=df, hue='Κατηγορία')

plt.title('Συνολικά Έσοδα ανά Προϊόν', fontsize=15)
plt.xlabel('Όνομα Προϊόντος', fontsize=12)
plt.ylabel('Έσοδα (€)', fontsize=12)
plt.xticks(rotation=45) # Πλάγια γράμματα για να διαβάζονται
plt.show()
```

### 2️⃣ Σχέση Τιμής και Ποσότητας (Scatter Plot)
Μήπως τα ακριβά προϊόντα πουλιούνται λιγότερο;

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Τιμή', y='Ποσότητα', data=df, s=100, color='red', alpha=0.7)

plt.title('Σχέση Τιμής - Ποσότητας Πωλήσεων')
plt.grid(True)
plt.show()
```

### 3️⃣ Κατανομή Εσόδων (Box Plot)
Το Box Plot μας βοηθάει να δούμε τη διασπορά των τιμών και αν υπάρχουν 'περίεργες' τιμές (outliers).

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Κατηγορία', y='Συνολικό_Έσοδο', data=df, palette='Set2')

plt.title('Διασπορά Εσόδων ανά Κατηγορία')
plt.show()
```
