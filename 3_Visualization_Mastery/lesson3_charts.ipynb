# 🎨 Lesson 3: Visualizing Trends
Σήμερα θα ομορφύνουμε τα δεδομένα μας!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Φόρτωση δεδομένων από το προηγούμενο μάθημα
df = pd.read_csv('../02_Datasets/sales_data.csv')
df['Συνολικό_Έσοδο'] = df['Τιμή'] * df['Ποσότητα']

# 1. Εντυπωσιακό Bar Chart με Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='Προϊόν', y='Συνολικό_Έσοδο', data=df, palette='viridis')
plt.title('Έσοδα ανά Προϊόν (Επαγγελματικό Style)')
plt.show()
```

Όπως βλέπετε, η Seaborn προσθέτει αυτόματα χρώματα και καλύτερη μορφοποίηση με πολύ λιγότερο κώδικα από την απλή Matplotlib.
