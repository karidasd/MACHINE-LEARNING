# 📈 Μάθημα 2: Ανάλυση Πραγματικών Δεδομένων
Σήμερα θα μάθουμε πώς να φορτώνουμε ένα αρχείο CSV και να απαντάμε σε ερωτήσεις όπως: *"Ποια κατηγορία είχε τα περισσότερα έσοδα;"*

```python
import pandas as pd

# Φόρτωση των δεδομένων από το GitHub (χρησιμοποιούμε το raw link)
url = '../02_Datasets/sales_data.csv'
df = pd.read_csv(url)

# Υπολογισμός Συνολικών Εσόδων ανά γραμμή
df['Συνολικό_Έσοδο'] = df['Τιμή'] * df['Ποσότητα']

print("Τα δεδομένα μας φορτώθηκαν!")
display(df.head())
```

### 📊 Ομαδοποίηση (Grouping)
Ας δούμε τα έσοδα ανά κατηγορία προϊόντων.

```python
category_revenue = df.groupby('Κατηγορία')['Συνολικό_Έσοδο'].sum()
print("Έσοδα ανά Κατηγορία:")
print(category_revenue)

category_revenue.plot(kind='pie', autopct='%1.1f%%', title='Κατανομή Εσόδων')
import matplotlib.pyplot as plt
plt.show()
```
