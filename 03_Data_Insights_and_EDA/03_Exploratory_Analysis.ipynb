# 🕵️‍♂️ Workshop: Exploratory Data Analysis (EDA)
Σήμερα θα ανακαλύψουμε κρυμμένα μοτίβα στα δεδομένα μας.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Φόρτωση και προετοιμασία
df = pd.read_csv('../02_Datasets/sales_data.csv')
df['Συνολικό_Έσοδο'] = df['Τιμή'] * df['Ποσότητα']

print("Στατιστική περίληψη:")
display(df.describe())
```

### 🌡️ Heatmap Συσχέτισης
Πόσο επηρεάζει η Τιμή την Ποσότητα; Οι τιμές κοντά στο 1 ή -1 δείχνουν δυνατή σχέση.

```python
plt.figure(figsize=(8,6))
# Επιλέγουμε μόνο τις αριθμητικές στήλες για τη συσχέτιση
corr = df[['Τιμή', 'Ποσότητα', 'Συνολικό_Έσοδο']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Πίνακας Συσχέτισης (Correlation Matrix)')
plt.show()
```

### 📊 Κατανομή Πωλήσεων
Ας δούμε πώς κατανέμονται οι τιμές των προϊόντων μας.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Τιμή'], kde=True, color='purple')
plt.title('Κατανομή Τιμών Προϊόντων (Histogram + KDE)')
plt.show()
```
